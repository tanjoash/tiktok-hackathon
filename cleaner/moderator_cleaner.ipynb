```python
# import libraries
import pandas as pd
import ast
```

```python
# read moderator_dimension.csv
mod = pd.read_csv("./moderator_dimension.csv")
mod.head()
```

```python
# get unique values in market into a list
vals = list(set(mod["market"]))
unique_vals = set()
for item in vals:
    values = [v.strip(' "[]') for v in item.split(',')]
    unique_vals.update(values)
unique_values_list = sorted(list(unique_vals))
print(unique_values_list)
```

```python
# convert string of list to list for column market
mod["market"] = mod["market"].apply(ast.literal_eval)
mod.head()
```

```python
# add columns in unique_values_list
for i in unique_values_list:
    mod[i] = 0
mod.head()
```

```python
# iterate through rows and update the corresponding cells
for index, row in mod.iterrows():
    for value in row['market']:
        mod.at[index, value] = 1

# drop "market column"
mod.drop(columns=["market"], inplace = True)
mod.head()
```

```python
mod = mod.dropna(axis = 0, subset = ["Productivity", "Utilisation %"])
```

```python
mod.rename(columns={'  accuracy  ': 'accuracy'}, inplace=True)
```

```python
# drop -
mod.drop(mod[mod["accuracy"] == "                  -   "].index, inplace = True)
```

```python
mod.to_csv("./cleaned_moderator_dimension.csv", index = False)
```
